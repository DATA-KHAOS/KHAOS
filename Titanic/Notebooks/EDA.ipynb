```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# VariÃ¡veis categÃ³ricas
cols_categoricas = ['pclass', 'embarked_c', 'embarked_q', 'embarked_s', 'age_group', 
                    'is_male', 'cabin_group', 'title', 'fare_group', 'is_alone', 
                    'sex_pclass_interaction']

for col in cols_categoricas:
    plt.figure(figsize=(8, 5))
    sns.countplot(x=col, hue='survived', data=df, palette='viridis')
    plt.title(f'SobrevivÃªncia por {col}')
    plt.show()
```

## ConclusÃ£o
* `pclass`: Pessoas na primeira classe tinham mais chances de sobreviver.
* `embarked`: O porto C foi o Ãºnico com maior chances de sobrevivÃªncia. 
* `age_group`: Claramente crianÃ§as foram retiradas primeiro.
* `is_male`: Mulheres tambÃ©m foram retiradas primeiro.
* `cabin_group`: Pessoas com os grupos C, E, D, B e F de cabine foram mais salvas do que nÃ£o.
* `title`: TÃ­tulos femininos foram mais salvos, reforÃ§ando a polÃ­tica de mulheres e crianÃ§as primeiro.
* `fare_group`: Quanto mais alta a taxa mais sobreviveram.
* `is_alone`: Quem esatava sozinho morreu mais do que quem nÃ£o estava.
* `sex_pclass_interaction`: Foram salvos pela ordem de mulheres da 1, 2 e 3 classe, em seguida de homens da 1, 3 e 2 classe. NÃ£o tendo um padrÃ£o tÃ£o claro em relaÃ§Ã£o aos homens. 

```python
# VariÃ¡veis numÃ©ricas
cols_numericas = ['age', 'sibsp', 'parch', 'fare', 'family_size', 'ticket_length', 'ticket_qtd']

for col in cols_numericas:
    plt.figure(figsize=(8, 5))
    sns.kdeplot(data=df[df['survived'] == 0][col], label='NÃ£o Sobreviveu', shade=True)
    sns.kdeplot(data=df[df['survived'] == 1][col], label='Sobreviveu', shade=True)
    plt.title(f'DistribuiÃ§Ã£o de {col} por sobrevivÃªncia')
    plt.legend()
    plt.show()
```
